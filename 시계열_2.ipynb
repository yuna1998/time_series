```r
firm <- c("Miler Lite", "Pepsi", "Storh's","Federal Express", "Burger King",
        "Coca-Cola", "McDonald's", "MCI", "Diet Cola", "Ford", "Levi's", "Bud Lite",
        "ATT Bell", "Calvin Klein", "Wendy's", "Polaroid", "Shasta", "Meow Mix",
        "Oscar Meyer", "Crest", "Kibbles N Bits")
```

```r
amount_spent <- c(50.1, 74.1, 19.3, 22.9, 82.4, 40.1, 185.9, 26.9,
                20.4, 166.2, 27, 45.6, 154.9, 5, 49.7,26.9, 5.7,
                7.6, 9.2, 32.4, 6.1 )
```

```r
Retained_Impression <- c(32.1, 99.6, 11.7, 21.9, 60.8, 78.6, 92.4, 50.7,
                        21.4, 40.1, 40.8, 10.4, 88.9, 12, 29.2, 38, 10, 12.3,
                        23.4, 71.1, 4.4)
```

```r
data <- data.frame(firm, amount_spent, Retained_Impression)
```

```r
head(data)
```

```r
model <- lm(amount_spent ~ Retained_Impression, data = data)
model
```

```r
summary(model)
```

```r
plot(model)
```

```r
confint(model)
```

```r
predict(model, interval = c("confidence"))[8,]
```

```r
predict(model, interval = c("prediction"))[8,]
```
